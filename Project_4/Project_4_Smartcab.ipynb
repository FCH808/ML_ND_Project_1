```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
summary_data = pd.read_csv('summary_info_1.csv')
```

```python
summary_data['reward_goal_ratio'] = summary_data.total_reward / summary_data.n_goal_reached
# If division by 0 (no goals reached), 
summary_data['reward_goal_ratio'] = summary_data['reward_goal_ratio'].replace('inf', 0)
```

```python
g = sns.factorplot(x='Q_val_starting_val', y='neg_score',
                   hue='replay_mem_ep',
                   data=summary_data,
                   size=12, legend_out=False, ci=None)
g.set_axis_labels('Starting Q-Values', 'Cumulative Negative Reward')
g.add_legend(title='Past Experiences\n Replayed per Update', label_order=['0','1','5','10','20','50'])
plt.title('Cumulative Negative Rewards by\n Starting Q-Value and Past Experiences Replayed per Update',
          size=14)
plt.figtext(.5,.06,'(Averaged over 10 Runs of 100 Trials Each)', fontsize=13, ha='center')

plt.savefig('neg_rewards.png', bbox_inches='tight')
```

```python
g = sns.factorplot(x='Q_val_starting_val', y='total_reward',
                   hue='replay_mem_ep',
                   data=summary_data,
                   size=12, legend_out=False)
g.set_axis_labels('Starting Q-Values', 'Cumulative Reward')
g.add_legend(title='Past Experiences\n Replayed per Update', label_order=['0','1','5','10','20','50'])
plt.title('Cumulative Rewards by\n Starting Q-Value and Past Experiences Replayed per Update',
          size=14)
plt.figtext(.5,.06,'(Averaged over 10 Runs of 100 Trials Each)', fontsize=13, ha='center')

plt.savefig('cum_rewards.png', bbox_inches='tight')
```

```python
g = sns.factorplot(x='Q_val_starting_val', y='total_updates',
                   hue='replay_mem_ep',
                   data=summary_data,
                   size=12, legend_out=False)
g.set_axis_labels('Starting Q-Values', 'Total Updates')
g.add_legend(title='Past Experiences\n Replayed per Update', label_order=['0','1','5','10','20','50'])
plt.title('Average Total Updates Applied by Starting Q-Value\n and Past Experiences Replayed per Update',
          size=14)
plt.figtext(.5,.06,'(Averaged over 10 Runs of 100 Trials Each)', fontsize=13, ha='center')

plt.savefig('total_updates.png', bbox_inches='tight')
```

```python
g = sns.factorplot(x='Q_val_starting_val', y='n_goal_reached',
                   hue='replay_mem_ep',
                   data=summary_data,
                   size=12, legend_out=False)
g.set_axis_labels('Starting Q-Values', 'Goals Reached')
g.add_legend(title='Past Experiences\n Replayed per Update', label_order=['0','1','5','10','20','50'])
plt.title('Average Goals Reached by Starting Q-Value\n and Past Experiences Replayed per Update',
          size=14)
g.set(ylim=(0, 100))
plt.figtext(.5,.06,'(Averaged over 10 Runs of 100 Trials Each)', fontsize=13, ha='center')

plt.savefig('goals_reached.png', bbox_inches='tight')
```

```python
g = sns.factorplot(x='Q_val_starting_val', y='reward_goal_ratio',
                   hue='replay_mem_ep',
                   data=summary_data,
                   size=12, legend_out=False, ci=None)
g.set_axis_labels('Starting Q-Values', 'Reward per Goal Reached')
g.add_legend(title='Past Experiences\n Replayed per Update', label_order=['0','1','5','10','20','50'])
plt.title('Average Reward per Goal Reached by Starting Q-Value\n and Past Experiences Replayed per Update',
          size=14)
plt.figtext(.5,.06,'(Averaged over 10 Runs of 100 Trials Each)', fontsize=13, ha='center')

plt.savefig('reward_per_goal_outlier.png', bbox_inches='tight')
```
